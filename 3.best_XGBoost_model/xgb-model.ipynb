```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import seaborn as sns
from sklearn.metrics import mean_absolute_error, r2_score
%matplotlib inline
import os
import pickle
from sklearn.metrics import r2_score
import xgboost as xgb
```

```python
with open('./Input/gap_29.pkl', 'rb') as file:
    x = pickle.load(file)

with open('./Input/EnergyGap.pkl', 'rb') as file:
    y = pickle.load(file)
# 划分数据集为训练集、验证集和测试集
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size=0.25, random_state=42)
model = xgb.XGBRegressor(gamma= 0.000123129, learning_rate=0.032845274,max_depth=8,min_child_weight= 2,n_estimators= 293, random_state=42)
model.fit(x_train, y_train)

y_train_pred = model.predict(x_train)
r2_train = r2_score(y_train, y_train_pred)
mae_train = mean_absolute_error(y_train, y_train_pred)

y_test_pred = model.predict(x_test)
r2_test = r2_score(y_test, y_test_pred)
mae_test = mean_absolute_error(y_test, y_test_pred)

y_val_pred = model.predict(x_val)
r2_val = r2_score(y_val, y_val_pred)
mae_val = mean_absolute_error(y_val, y_val_pred)

print("最终模型")
print("训练集 R2:", r2_train)
print("测试集 R2:", r2_test)
print("验证集 R2:", r2_val)
print("训练集MAE", mae_train)
print("测试集MAE", mae_test)
print("验证集MAE", mae_val)
```

```python
# =========================================================================
# 5-fold cross-validation
# =========================================================================
from sklearn.model_selection import cross_validate, KFold

print("\n" + "="*40)
print("正在进行 5折交叉验证以评估模型稳定性...")

best_params = {
    'gamma': 0.000123129,
    'learning_rate': 0.032845274,
    'max_depth': 8,
    'min_child_weight': 2,
    'n_estimators': 293,
    'random_state': 42,
    'n_jobs': -1  
}


cv_model = xgb.XGBRegressor(**best_params)

kf = KFold(n_splits=5, shuffle=True, random_state=42)


scoring_metrics = {
    'r2': 'r2',
    'neg_mae': 'neg_mean_absolute_error' 
}


import numpy as np
y_cv = np.array(y).ravel() 

cv_results = cross_validate(
    cv_model, 
    x, y_cv,      
    cv=kf, 
    scoring=scoring_metrics,
    return_train_score=False 
)

cv_r2_scores = cv_results['test_r2']
cv_mae_scores = -cv_results['test_neg_mae'] 


print(f"\n[5-Fold CV 详细结果]")
print(f"每折 R2: {cv_r2_scores}")
print(f"每折 MAE: {cv_mae_scores}")

print("-" * 40)

print(f"Average R2 : {np.mean(cv_r2_scores):.4f} ± {np.std(cv_r2_scores):.4f}")
print(f"Average MAE: {np.mean(cv_mae_scores):.4f} ± {np.std(cv_mae_scores):.4f}")
print("="*40)


cv_r2_scores = cv_results['test_r2']

cv_mae_scores = -cv_results['test_neg_mae'] 

# ==========================================
# English Output Section (Ready for Manuscript)
# ==========================================
print(f"\n[5-Fold Cross-Validation Detailed Results]")
print(f"R2 scores per fold:  {cv_r2_scores}")
print(f"MAE scores per fold: {cv_mae_scores}")

print("-" * 50)
print(f"Final Statistical Summary:")

print(f"Average R2 : {np.mean(cv_r2_scores):.4f} ± {np.std(cv_r2_scores):.4f}")
print(f"Average MAE: {np.mean(cv_mae_scores):.4f} ± {np.std(cv_mae_scores):.4f}")
print("=" * 50)
```
